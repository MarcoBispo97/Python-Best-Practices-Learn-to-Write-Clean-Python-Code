## Implied Line Continuation in Python
- The recommended way to wrap long lines in Python code
- Surround the line(s) with parentheses(), brackets[], or braces{}.

````python
password = "pythonisawesome"

if (len(password) >= 8 
        and any (letter.isupper() for letter in password)
        and any(letter.islower() for letter in password) 
        and any(letter.isdigit() for letter in password)):
    print(f"The password {password} is valid")
else:
    print(f"The password {password} is invalid")
````

````python
def format_string(string):
    words = string.split(" ")
    return (words[0].upper()
            + " " 
            + " ".join(words[1:-1]).lower() 
            + " "
            + " " + words[-1].upper())
````

## Line Breaks with Backslashes
- Avoid using backslashes (\) to break long lines in Python code
- Implicit vs Explicit line continuation
- \ backslashes can lead to errors if not used carefully
- / forward slashes can be overlooked or forgotten, leading to syntax errors
- Implicit line continuation should be used in preference to using a backslash

````python
price = 50
tax = 5
discount = 1
coupons = 2

total_price = price + \
                tax - \
                discount - \
                coupons

total_price = (price
                + tax
                - discount
                - coupons)
````

## Binary Operators in a Line Break
- An operator that operates on two operands (values) and manipulates them to return a result.
- Arithmetic operators: +, -, *, /, %, **, //
- Comparison operators: ==, !=, >, <, >=, <=
- Logical operators: and, or, not
- Bitwise operators: &, |, ^, ~, <<, >>
- When breaking a line, place the binary operator at the beginning of the new line.

## Compound Statements

- Statement that contains a group of other statements.
- A compound statement consists of one or more 'clauses'. Each clause consists of a header and a 'suite'.

Compound Statements in Python include:
- Conditions
- While Loops
- For Loops
- Try Statement
- with Statement

ğŸª Best Practice #1

```python

for i in range(10):
    do_something()
``` 

## Blank Lines
- Used to group related code blocks or logical sections.
- ğŸª Surround top-level function and class definitions with `two` blank lines.
```python
def function_one():
    pass


def function_two():
    pass


def function_three():
    pass
```


- Surround method definitions inside a class with a `single` blank line
```python
class MyClass:
    #1
    def method_one(self):
        pass
    #1
    def method_two(self):
        pass
    #1
    #2
class AnotherClass:
    #1
    def another_method(self):
        pass
```


- ğŸª Use additional blank lines (sparinggly) to separate groups of related functions or methods.
```python
class Player:
    def __init__(self, name):
        self.x = x
        self.y = y
    #1
    #Move vertically
    def move_up(self):
        self.y += 1

    def move_down(self):
        self.y -= 1
    #1
    #2
    #Move horizontally
    def move_left(self):
        self.x -= 1
    #1
    def move_right(self):
        self.x += 1
```

- ğŸª Use blank lines in functions (sparinggly) to indicate logical sections.
```python
def is_prime(number):
    if number <= 1:
        return False
    #1
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    #1
    return True
```
