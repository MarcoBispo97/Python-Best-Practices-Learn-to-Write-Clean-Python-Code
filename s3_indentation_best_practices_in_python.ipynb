# Identantion Best Practices in Python

## Importance of Identation

- It is the spacing of code to indicat its logical structure
- It is used to group related lines of code and to make code more readable and understandable

## Identation Guidelines (Tabs vs Spaces)


- Use 4 spaces per level of identation

### No extra indentation

````python
if (this_is_one_thing and
    that_is_another_thing):
    do_something()
````

### Add a comment, wich will provide some distinction in editors
### supporting syntax highlighting.


````python
if (this_is_one_thing and
    that_is_another_thing):
    # Since both conditions are true, we can frobnicate.
    do_something()
````

### Add some extra identation on the conditional continuation line.


````python
if (this_is_one_line 
        and
    that_is_on_another_line):
    do_something()
````
